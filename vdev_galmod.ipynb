Find the min and max of the VLSR allowed by Galactic rotation curve at a given (l, b). <br>
b in step of 0.5, l in step of 0.5; could be changed accordingly in the find_vlsr_minmax_allsky function. 

```python
from find_vlsr_minmax_allsky import find_vlsr_minmax_allsky
b, l, vmin_allsky, vmax_allsky = find_vlsr_minmax_allsky()
```

Then, for example, the Vlsr when Vdev = +/-20 will be: 

```python
vlsr_vdev_pos20 = vmax_allsky+20
vlsr_vdev_neg20 = vmin_allsky-20
```

```python
# plot the curve
import matplotlib.pyplot as plt
```

plot some Vdev curves at b=0

```python
# plot the curve
# find the curves at b = 0
ind_b0 = b==0.

# vdev = 0 curve
vmin_b0_vdev0 = vmin_allsky[ind_b0, :].flatten()
vmax_b0_vdev0 = vmax_allsky[ind_b0, :].flatten()

# vdev = +/- 60 curve
vmin_b0_vdev60 = vmin_allsky[ind_b0, :].flatten()-60
vmax_b0_vdev60 = vmax_allsky[ind_b0, :].flatten()+60
```

```python
fig = plt.figure(figsize=(6, 4))
ax = fig.add_subplot(111)

# the vdev = 0
ax.plot(l, vmin_b0_vdev0, color='k', linestyle='-', label='Vdev = 0', lw=2)
ax.plot(l, vmax_b0_vdev0, color='k', linestyle='-', lw=2)

# the vdev = +/-20 curve
ax.plot(l, vmin_b0_vdev60, color='r', linestyle='--', label='Vdev=+/-60', lw=2)
ax.plot(l, vmax_b0_vdev60, color='r', linestyle='--', lw=2)

lg = ax.legend(loc='lower right')

ax.minorticks_on()
ax.set_xlim(360, 0)
ax.set_ylim(-250, 250)
ax.set_xlabel('Glon (degree)')
ax.set_ylabel('Velocity (km/s)')
ax.set_title('b=0')
```

Now, plot some curves at b = 0, 15, 40 and vdev=0 <br>
This plot can be found in http://adsabs.harvard.edu/abs/1991A%26A...250..499W 

```python
# plot the curve

# b = 15 curve
ind_b15 = b==15.
vmin_b15_vdev0 = vmin_allsky[ind_b15, :].flatten()
vmax_b15_vdev0 = vmax_allsky[ind_b15, :].flatten()

# v = 40 curve
ind_b40 = b==40.
vmin_b40_vdev0 = vmin_allsky[ind_b40, :].flatten()
vmax_b40_vdev0 = vmax_allsky[ind_b40, :].flatten()
```

```python
fig = plt.figure(figsize=(6, 4))
ax = fig.add_subplot(111)

# the b=0
ax.plot(l, vmin_b0_vdev0, color='k', linestyle='-', label='b=0', lw=2)
ax.plot(l, vmax_b0_vdev0, color='k', linestyle='-', lw=2)

# the b=15
ax.plot(l, vmin_b15_vdev0, color='r', linestyle='--', label='b=15', lw=2)
ax.plot(l, vmax_b15_vdev0, color='r', linestyle='--', lw=2)

# the b=40
ax.plot(l, vmin_b40_vdev0, color='b', linestyle=':', label='b=40', lw=2)
ax.plot(l, vmax_b40_vdev0, color='b', linestyle=':', lw=2)

lg = ax.legend(loc='lower right')

ax.minorticks_on()
ax.set_xlim(360, 0)
ax.set_ylim(-250, 250)
ax.set_xlabel('Glon (degree)')
ax.set_ylabel('Velocity (km/s)')
ax.set_title('Vdev=0')
```
